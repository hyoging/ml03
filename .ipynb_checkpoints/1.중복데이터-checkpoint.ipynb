```python
import numpy as np
nan_list = ['대여거치대', '반납거치대','이용시간','이용거리']
for number in df['대여소번호'].unique():
    for name in nan_list:
        mean = round(df[name][df['대여소번호']==number].mean(),1)
        df.loc[df['대여소번호'] == number, name] = df[name][df['대여소번호']== number].name
        

df.loc[3]
        
```

```python
import numpy as np
q1, q3 = np.percentile(df['나이'],[25,75])
iqr = q3 - q1
upperfence = q3 + iqr * 1.5
print('iqr * 1.5 : ' , upperfence)

df[df['나이']>upperfence]
```

```python
upperfence = q3 + iqr*3
print(upperfence)
df[df['나이'] > upperfence]
```

```python
df[df['나이'] > upperfence] = df['나이'].median()
```
